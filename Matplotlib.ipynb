```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

# Charts :

## 1. Pie Chart :

```python
def batsman_record(name):
    
    new_delivery=delivery[delivery['batsman']==name]
    x=new_delivery['batsman_runs'].value_counts().sort_values(ascending=False)
    plt.pie(x, labels=x.index, autopct="%0.2f%%")
    plt.show()
```

```python
batsman_record("V Kohli")
```

## 2. Line Chart :

```python
big=match.merge(delivery,left_on='id',right_on='match_id')
batsman_data=big[big['batsman']=='V Kohli']
performance=batsman_data.groupby('season').sum()['batsman_runs']

plt.plot(performance.index , performance.values)
plt.xlabel('season')
plt.ylabel('Runs Scored')
```

```python
def performance_comparing(name1 , name2):
    big=match.merge(delivery,left_on='id',right_on='match_id')
    batsman_data=big[big['batsman']==name1]
    batsman_data1=big[big['batsman']==name2]
    performance=batsman_data.groupby('season').sum()['batsman_runs']
    performance1=batsman_data1.groupby('season').sum()['batsman_runs']

    plt.plot(performance.index , performance.values, color="orange", marker='+', label=name1)
    plt.plot(performance1.index , performance1.values, color="green", marker='d', label=name2)
    plt.xlabel('season')
    plt.ylabel('Runs Scored')
    plt.title('Batsman season by season performance')
    plt.grid()
    plt.legend()

nam1=input("enter a batsman name : ")
nam2=input("enter another batsman name : ")
performance_comparing(nam1,nam2)
```

```python
plt.scatter(performance.index,performance.values)
```

## 3. Bar Chart :

```python
new_delivery=delivery[delivery['batsman']=="V Kohli"]
x=new_delivery['batsman_runs'].value_counts().sort_values(ascending=False)
plt.bar(x.index,x.values)
plt.show()
```

## 4. Histogram

```python
runs=delivery.groupby('batsman').sum()['batsman_runs']
balls=delivery['batsman'].value_counts()
sr=(runs/balls)*100
sr
plt.hist(sr.values, bins = 100)
plt.show()
```
